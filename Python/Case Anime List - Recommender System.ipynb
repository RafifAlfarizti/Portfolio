# Data Prepraration

```python
an = pd.read_csv(r"D:\Bootcamp data scientist\Final Project\Dataset Kaggle\animes.csv\animes.csv")
an
```

```python
a = []
b = []
c = []
for x, y, z in zip(an['uid'], an['title'], an.index):
    if (x not in a) & (y not in b):
        a.append(x)
        b.append(y)
    if (x not in a) & (y in b):
        c.append(z)
```

Keterangan kode:
1. zip(an['uid'], an['title'], an.index) menggabungkan tiga urutan berbeda (kolom 'uid', kolom 'title', dan indeks) dari objek an.
2. a, b, dan c diinisialisasi sebagai daftar kosong.
3. for kemudian mengiterasi melalui elemen-elemen gabungan ini, mengekstrak tiga nilai sekaligus (x, y, dan z).

Kondisi Pertama: <br>
if (x not in a) & (y not in b):<br>
Memeriksa apakah nilai x (dari kolom 'uid') belum ada dalam daftar a dan nilai y (dari kolom 'title') belum ada dalam daftar b. <br>
Jika kondisi ini benar, maka:<br>
a.append(x): Menambahkan x ke daftar a. <br>
b.append(y): Menambahkan y ke daftar b. <br>

Kondisi Kedua: <br>
if (x not in a) & (y in b): <br>
Memeriksa apakah nilai x belum ada dalam daftar a, namun nilai y sudah ada dalam daftar b. <br>
Jika kondisi ini benar, maka: <br>
c.append(z): Menambahkan z (indeks dari elemen saat ini dalam objek an) ke daftar c. <br>

# Memberi Peringkat Anime Terbaik 

```python
# Normalisasi kolom 'members' dan 'score' untuk menggabungkan dalam satu metrik
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
an['normalized_members'] = scaler.fit_transform(an[['members']])
an['normalized_score'] = scaler.fit_transform(an[['score']])

# Menghitung 'popularity_score' berdasarkan beberapa faktor
an['popularity_score'] = an['normalized_members'] * 0.6 + an['normalized_score'] * 0.4

# Mengurutkan DataFrame berdasarkan 'popularity_score'
top_anime = an.sort_values(by='popularity_score', ascending=False)

# Menampilkan 10 anime paling populer berdasarkan 'popularity_score'
top_anime.head(10)
```
