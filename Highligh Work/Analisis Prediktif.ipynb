```python
# Impor library yang diperlukan
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Membaca data dari file Excel
data = pd.read_excel("E:\Bahan Statistika\IHSG.xlsx")
data
```

```python
# Membuat plot untuk melihat tren penjualan seiring waktu
plt.figure(figsize=(10, 5))
plt.plot(data['Period'], data['IHSG'], marker='o')
plt.title('Tren IHSG')
plt.xlabel('Period')
plt.ylabel('IHSG')
plt.xticks(rotation=45)
plt.show()
```

```python
# Membuat model regresi linier untuk memprediksi pendapatan berdasarkan produk terjual
model = LinearRegression()
X = data[['USD']]
y = data['IHSG']
model.fit(X, y)
```

```python
# Menampilkan koefisien regresi
print("Koefisien Regresi:", model.coef_[0])
```

```python
# Prediksi pendapatan untuk 200 produk terjual
USD_prediksi = 15000
IHSG_prediksi = model.predict([[USD_prediksi]])
print("Prediksi IHSG jika nilai 1 USD adalah 15000:", IHSG_prediksi[0])
```
