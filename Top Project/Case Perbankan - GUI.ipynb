```python
import pandas as pd
import numpy as np
from tkinter import *
from matplotlib.figure import Figure
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

# Menghasilkan data perbankan acak
np.random.seed(0)
jumlah_nasabah = 1000
nama_nasabah = ['Nasabah_' + str(i) for i in range(jumlah_nasabah)]
saldo_awal = np.random.uniform(1000, 100000, jumlah_nasabah)
transaksi = np.random.uniform(-5000, 5000, jumlah_nasabah)
saldo_akhir = saldo_awal + transaksi

data_perbankan = pd.DataFrame({
    'Nama Nasabah': nama_nasabah,
    'Saldo Awal': saldo_awal,
    'Transaksi': transaksi,
    'Saldo Akhir': saldo_akhir
})

# Analisis data
rata_saldo_awal = data_perbankan['Saldo Awal'].mean()
rata_transaksi = data_perbankan['Transaksi'].mean()
rata_saldo_akhir = data_perbankan['Saldo Akhir'].mean()

# Membuat GUI
root = Tk()
root.title("Analisis Data Perbankan")

# Membuat frame untuk tabel data
frame = Frame(root)
frame.pack(side=TOP, fill=BOTH, expand=True)

# Menampilkan data dalam bentuk tabel
table = Text(frame, height=10, width=100)
table.pack(side=LEFT, fill=BOTH, expand=True)
table.insert(END, data_perbankan.to_string(index=False))

# Membuat frame untuk analisis
analysis_frame = Frame(root)
analysis_frame.pack(side=TOP, fill=BOTH, expand=True)

# Menampilkan hasil analisis
analysis_label = Label(analysis_frame, text=f"Rata-rata Saldo Awal: {rata_saldo_awal:.2f}\nRata-rata Transaksi: {rata_transaksi:.2f}\nRata-rata Saldo Akhir: {rata_saldo_akhir:.2f}", justify=LEFT)
analysis_label.pack(side=LEFT, padx=10)

# Membuat grafik distribusi saldo akhir
fig = Figure(figsize=(5, 4), dpi=100)
plot = fig.add_subplot(111)
plot.hist(data_perbankan['Saldo Akhir'], bins=50, color='green', alpha=0.7)
plot.set_title('Distribusi Saldo Akhir Nasabah')
plot.set_xlabel('Saldo Akhir')
plot.set_ylabel('Jumlah Nasabah')

# Menambahkan grafik ke GUI
canvas = FigureCanvasTkAgg(fig, master=root)
canvas.draw()
canvas.get_tk_widget().pack(side=TOP, fill=BOTH, expand=True)

root.mainloop()
```
